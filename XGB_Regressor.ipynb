```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import mean_squared_error
from xgboost import XGBRegressor
```

```python
data = {
    "Bedrooms": [3, 2, 4, 3, 5, 3, 2, 4],
    "Square_Feet": [1500, 1200, 2500, 1800, 3000, 2500, 1800, 3000],
    "Age": [10, 20, 15, 5, 8, 20, 15, 5],
    "Price": [300000, 200000, 400000, 350000, 500000, 300000, 200000, 400000]
}
df = pd.DataFrame(data)
```

```python
X = df.drop('Price',axis=1)
y = df['Price']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.1,random_state=42)
```

```python
model = XGBRegressor()
```

```python
parameters = {
    'max_depth' : [2,3],
    'learning_rate':[0.2,0.6,1.0],
    'n_estimators' :[50,80],
    'objective' : ['reg:squarederror']
}
```

```python
grid = GridSearchCV(model,parameters,cv=5)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
xgb = XGBRegressor(learning_rate = 0.2,
 max_depth= 2,
 n_estimators = 50,
 objective = 'reg:squarederror')
```

```python
xgb.fit(X_train,y_train)
y_pred = xgb.predict(X_test)
```

```python
mean_squared_error(y_test,y_pred)
```
