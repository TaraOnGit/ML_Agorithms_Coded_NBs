```python
import tensorflow as tf
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
img = image.load_img('/content/cat1.png',target_size=(200,200))
```

```python
img
```

```python
datagen = ImageDataGenerator(
    rotation_range=30,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    width_shift_range=0.2,
    height_shift_range=0.2,
    fill_mode='nearest'

)
```

```python
image_arr = image.img_to_array(img)
```

```python
image_arr.shape
```

```python
input_batch = image_arr.reshape(1,200,200,3)
```

```python
!mkdir pics
```

```python
i = 0

for output in datagen.flow(input_batch,batch_size=1,save_to_dir='pics'):
  i = i + 1
  if i == 10:
    break
```
