```python
from sklearn.datasets import make_classification
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
X,y = make_classification(n_samples=1000,n_features=2, n_informative=1,n_redundant=0,
                           n_classes=2, n_clusters_per_class=1,random_state=42)
```

```python
plt.figure(figsize=(10,10))
plt.scatter(X[:,0],X[:,1],cmap='winter',c=y)
plt.show()
```

```python
def perceptron(X,y):
  X = np.insert(X,0,1,axis=1)
  weights = np.ones(X.shape[1])
  lr = 0.1

  for i in range(50000):
    j = np.random.randint(0,1000)
    y_hat = step(np.dot(X[j],weights))
    weights = weights + lr * (y[j] - y_hat) * X[j]

  return weights[0],weights[1:]
```

```python
def step(z):
 return 1 if z > 0 else 0
```

```python
intercept_, coef_ = perceptron(X,y)
```

```python
print(intercept_, '   ',coef_)
```

```python
m = - (coef_[0] / coef_[1])
b = - (intercept_ / coef_[1])
```

```python
X_input = np.linspace(-3,3,100)
y_input = m * X_input + b
```

```python
plt.figure(figsize=(10,6))
plt.plot(X_input,y_input,color='red',linewidth=3)
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
plt.ylim(-3,2)
```
